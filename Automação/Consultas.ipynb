WebScrapping do conteudo dos termos em PE.

```python
import pandas as pd
from selenium import webdriver
from time import sleep as slp
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.chrome.service import Service
import csv

df = pd.read_excel('C:\\Users\\ferna\\Downloads\\Pesquisa PE.xlsx', dtype=str, sheet_name='Termo')

servico = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=servico)
browser.maximize_window()

# Abrir a página da web
url = 'https://efisco.sefaz.pe.gov.br/sfi_trb_cmt/PRConsultarTermoFielDepositario'

# Abrir o arquivo CSV para escrita
filename = 'Log_PE.csv'
with open(filename, 'a', newline='', encoding='UTF-16') as arquivo:
    writer = csv.writer(arquivo)

    for row in df.itertuples():
        t = row.Termo
        
        # Abrir a página da web
        browser.get(url)
        
        # Preencher o campo de busca
        termo_input = browser.find_element(By.ID, 'nuTermoFielDepositario')
        termo_input.clear()
        termo_input.send_keys(t)
        
        # Clicar no botão de busca
        localizar_button = browser.find_element(By.ID, 'btt_localizar')
        localizar_button.click()
        
        # Clicar no botão de detalhar
        detalhar_button = browser.find_element(By.ID, 'btt_detalhar')
        detalhar_button.click()
        
        element_xpath = '//*[@id="nuDocumentoTransportadora"]'
        element = browser.find_element(By.XPATH, element_xpath)
        valor = element.get_attribute("value")
        
        # Extrair os dados da tabela e escrever no arquivo CSV
        table_id = "table_tabeladados"  # substitua pelo ID real do elemento da tabela
        table_element = browser.find_element(By.ID, table_id)
        rows = table_element.find_elements(By.TAG_NAME, "tr")

        for row in rows:
            cells = row.find_elements(By.TAG_NAME, "td")
            row_data = [cell.text for cell in cells]
            row_data.insert(0, valor)
            writer.writerow(row_data)

# Fechar o navegador
browser.quit()
```

WebScrapping do conteudo dos registros NFe em PE.

```python
import pandas as pd
from selenium import webdriver
from time import sleep as slp
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.chrome.service import Service
import csv

df = pd.read_excel('C:\\Users\\ferna\\Downloads\\Pesquisa PE.xlsx', dtype=str, sheet_name='TESTE')

servico = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=servico)
browser.maximize_window()

# Abrir a página da web
browser.get('https://efisco.sefaz.pe.gov.br/sfi_trb_cmt/PRConsultarNFTermoFielDepositario')
alert = Alert(browser)
alert.accept()

# Abrir o arquivo CSV para escrita
filename = 'Log_PE.csv'
with open(filename, 'a', newline='', encoding='UTF-16') as arquivo:

    for row in df.itertuples():
        t = row.Registro
        
        # Preencher o campo de busca
        termo_input = browser.find_element(By.XPATH, '//*[@id="primeiro_campo"]')
        termo_input.clear()
        termo_input.send_keys(t)
        
        # Clicar no botão de busca
        localizar_button = browser.find_element(By.XPATH, '//*[@id="btt_localizar"]')
        localizar_button.click()
        
        registro = browser.find_element(By.XPATH, '//*[@id="table_tabeladados"]/tbody/tr[2]')
        texto = registro.text
        
        arquivo.write(
            str(texto))
        arquivo.write(str('\n'))

# Fechar o navegador
browser.quit()
```
